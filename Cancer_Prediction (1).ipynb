### Data Preprocessing

##### Label encoding

##### Here, I assign the 30 features to a NumPy array X, and encoded original string representation (M and B) into integers to start machine learning phase.

```python
from sklearn.preprocessing import LabelEncoder

X=data.iloc[:,1:11].values
y=data.iloc[:,0].values

le=LabelEncoder()
y = le.fit_transform(y)
```

```python
y #(M=1,B=0)
```

```python
np.unique(y, return_counts=True)
```

###### the malignant tumors are now represented as 1 and the benign tumors are represented as 0 (i.e no cancer cells detection)

### Split data into training and test sets

```python
X_train, X_test,y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
X_train.shape,y_train.shape, X_test.shape, y_test.shape
```

### Feature Standardization(Scaling our data)

```python
scaler =StandardScaler()
Xs = scaler.fit_transform(X)
```

#### Predictive machine learning model using Support Vector Machine (SVM)

```python
#create classifier
clf=SVC(probability=True)
clf.fit(X_train, y_train)
```

```python
y_pred=clf.predict(X_test)
```

```python
y_pred
```

```python
svm_acc=accuracy_score(y_test,y_pred)
```

```python
print('\nThe classifier accuracy score is :',svm_acc)
```

### Model accuracy is 92.1%

### Confusion Matrix

```python
cm=confusion_matrix(y_test,y_pred)
```
